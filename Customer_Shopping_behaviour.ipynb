```python
df["Review Rating"]= df.groupby("Category")["Review Rating"].transform(lambda x:x.fillna(x.median()))
df["Review Rating"]
```

```python
df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(" ","_") # all the column names are in snake casing  we have converted it becuase  to avoide error we  when we'll do analysis in sql
df=df.rename(columns={"purchase_amount_(usd)" : "purchase_amount"})
```

```python
# Create a Colunms age_group
labels=["Young Adult","Adult","Middle-aged","Senior"]
df["age_group"]=pd.qcut(df["age"],q=4,labels=labels)
df[["age","age_group"]]
```

```python
# create column purchase_frequency_days 
freq_to_days = { "Daily": 1,"Weekly": 7,"Bi-Weekly": 14,"Fortnightly": 14,"Monthly": 30,"Quarterly": 90,"Annually": 365}
df["purchase_frequency_days"]=df["frequency_of_purchases"].map(freq_to_days)
df["purchase_frequency_days"]
```
